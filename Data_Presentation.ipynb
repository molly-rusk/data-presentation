```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df['Age'].value_counts().sort_values(ascending=False)
```

```python
df.groupby('Age')['Absenteeism time in hours'].sum().sort_values(ascending=False)
```

```python
absencesperage = df.groupby('Age')['Absenteeism time in hours'].sum()
absencesperage
```

```python
fig, ax = plt.subplots()
absencesperage.plot.bar(ax=ax, stacked=True)
plt.ylabel('Hours')
plt.title('Total hours of absence per age')
plt.xticks(rotation=60)
fig.tight_layout()
```

```python
ages = df['Age'].value_counts()
ages
```

```python
ages.plot.bar()
plt.title('Counts of ages in the workplace')
plt.ylabel('Count')
plt.xlabel('Age')
plt.show()
```

```python
dayabsent = df['Day of the week'].value_counts().head()
dayabsent
```

```python
fig, ax = plt.subplots()
dayabsent.plot.bar()
plt.xlabel('Day of the week')
plt.ylabel('Total days absent')
plt.title('Days of absence')
plt.xticks(rotation=0)
fig.tight_layout()
```

```python
monthofabsence = df['Month of absence'].value_counts()
monthofabsence
```

```python
monthofabsence.plot.bar()
plt.title('Number of absences per month')
plt.xticks(rotation=0)
plt.xlabel('Month')
plt.ylabel('Absences')
plt.show()
```
